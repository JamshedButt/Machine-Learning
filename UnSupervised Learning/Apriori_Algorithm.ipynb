```python
# Dummy Dataset
dataset = [
    ['Malik','Onion','Nutmeg','Kidney Beans','Eggs','Yougurt'],
    ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yougurt'],
    ['Malik','Apple','Kidney Beans','Eggs'],
    ['Malik','Unicon','Corn','Kidney Beans','Yougurt'],
    ['Corn','Onion','Onion','Ice Cream','Eggs'] 
]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder #(It is help for encoding)
```

```python
te = TransactionEncoder()
```

```python
te_array = te.fit(dataset).transform(dataset)
```

```python
te_array
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.6, use_colnames=True)
```
