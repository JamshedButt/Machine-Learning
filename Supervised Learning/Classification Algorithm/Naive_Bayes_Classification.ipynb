```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
df.head()
```

```python
df.shape
```

```python
#data Precessing

x = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
# Spliting The data
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print("train_data : ",train_data.shape)
print("train_label : ",train_label.shape)
print("test_data : ",test_data.shape)
print("test_label : ",test_label.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
train_data = sd.fit_transform(train_data)
test_data = sd.fit_transform(test_data)
```

```python
train_data
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(train_data,train_label)
```

```python
y_pred = model.predict(test_data)
```

```python
y_pred
```

```python
test_label
```

```python
#Check Accuracy

from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Model Accuracy : ",accuracy_score(test_label,y_pred))
```

```python
cm = confusion_matrix(test_label,y_pred)
```
