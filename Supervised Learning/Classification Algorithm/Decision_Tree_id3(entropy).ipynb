```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,[2,3]]
y = df.iloc[:,4]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
```

```python
train_data = sd.fit_transform(train_data)
test_data = sd.fit_transform(test_data)
```

```python
train_data
```

```python
test_data
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="entropy",random_state=0)
```

```python
model.fit(train_data,train_label)
```

```python
y_pred = model.predict(test_data)
```

```python
y_pred
```

```python
test_label
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(test_label,y_pred))
```

```python
cm = confusion_matrix(test_label,y_pred)
```

```python
print(cm)
```
